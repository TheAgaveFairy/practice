```python
df = pd.read_csv('BarTools26Jan22.csv')
df = df.dropna(how="all",axis=1)
#df.head()
```

```python
df = df.drop(columns=['Category','StockValue','VolumeOnHand','Available','OnOrder','InTransit','Allocated','Status','Unit','PriceTier1','PriceTier2','PriceTier3'])
df.head()
```

```python
pivot = df.pivot(index='ProductName',columns='Location',values='OnHand')
pivot.to_csv('BarToolPivot.csv')
pivot
```
